# **Analysis of the 6 common sentences across all sentences**
**Sentence 12**:  We can observe that the CTI highlighted the pronoun "dumneavoastră" in two instances, one for each sentence of the generated output. In each situation, the almost all highlighted contextual cues (determiners, nouns, word-parts) are not relevant to suggesting a formal pronoun, as generated. This outcome could be explained by the lack of usage of any second person pronouns in the context. However, certain words could imply the usage of a formal versus an informal form, but PECoRe usually fails to detect these tokens, except for the "Afa(0.001)cerilor"(Affairs). Although this token suggests formality, it has a very low score attributed to it, which is in line with the other tokens identified by the CCI step. This fact indicates the model is "coming to the right conclusion for the wrong reasons".

**Sentence 18**: In this example, we can observe a lack of contextual cues, the only ones being represented by the expression "în casă" (wrongly translated) and the conjunction și (and). None of these tokens are contributing to the generation of the context-sensitive target span represented by the verb "călătorești(0.667)" (travel) which implies the informal singular form of you by the suffix "-ști:. The informal form is also maintained for the rest of the sentence, suggested by the verb "așteaptă" (wait). Therefore, similar to the previous case, the model provides a correct translation, but does not rely on context for generating it.

**Sentence 25**: Similar to sentence 18, we can observe a shortage of contextual cues in this case. Moreover, we can observe that the only token highlighted by CTI is the verb "e(0.057)", which is the contracted form of the verb "este" (is), suggesting an informal setting. This is supported by the translation of  "you" to "tine" and the verb form "să ai" (to have). Nevertheless, the model does not manage to justify its decision even though the context contains the verb "a fi" (to be), which is specific to the informal singular form of you.


**Sentence 29**: This instance is the first one in which we can observe that the model rightfully justifies its generated output concerning the phenomenon we are interested in. Two verbs are highlighted by the CTI, "po(1.392)ți" (can) and "vei(0.332)" (will) that indirectly indicate the informal singular form of "you". In each case, the contextual cues are represented by the verbs that suggest the same pronoun form. Although the generated output contains further evidence for the informal singular form of "you" ("te-ai întors"), this is not detected by CTI. This might be due to the fact that it already captured the former instances and this later instance is dependent on them.

**Sentence 39**: Akin to the previous sentence, the informal singular form of you is indicated indirectly through the verb "condu(0.268)ci" (drive), which relies on the contextual cue "deci(0.003)zi" (decide), which indirectly suggests the informal singular form of "you". Other verb forms could have contributed to the generated output but are not captured by either CTI or CCI.

**Sentence 53**: In this case, CTI highlighted the verb "trimiteți(2.056)" (send) that suggests a formal version of the pronoun "you". However, all the highlighted contextual cues are not relevant to suggesting a formal pronoun, as generated. Among these contextual clues is the part-word "vă(0.003)muirea" (customs) which, taken separately as "vă" (you), could suggest the generated output. Regardless if the model interprets vă as you as indicative of the formal tone, it relies on incorrect context to derive its conclusion.
