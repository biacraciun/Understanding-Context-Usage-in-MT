# **Analysis of the 6 common sentences across all sentences**

**Sentence 12**: The CTI identified multiple tokens, the last one being the phenomena in question: the word “ви” which translates to “your” in reference to the noun “passport”. The highlighted CCI, however, are related to the words “requirement”, and “this” found in the context sentences and neither is directly related to the formality pronoun. The pronoun itself is the formal version of “your” which does not differentiate between genders, just formal/informal. It should be noted here that there are two ways in Bulgarian to say “your passport”: the first one is by using the literal word for “your” which is “Вашият”, and the second way is more indirect which is what we see in the translation from PECoRe as it does not directly use that word but a variation of it to refer to the noun before it. Additionally, the generated output sentence in Bulgarian from PECoRe slightly differs from the original input sentence in Bulgarian. PECoRe did not translate the second part of the utterance. For the first part, it is mostly the same, with the original sentence having indirect reference to “you” by the use of “имате”, meaning “you have” where the “you” is included at the end of the word “have”, while the generated sentence has “ви” meaning “your”. Both are referring to the noun “passport”, just translated differently, however, both are correct translations. In this example PECoRe did take into account some of the words from the context, but not the right words in relation to the formality pronoun and one reason for that could be the lack of “you” in the context itself.

**Sentence 18**: In this example the CTI identified 2 tokens, the second one being more relevant for this analysis as it highlights the words “you travel” or “пътувате” in Bulgarian. Similarly to the first example above, the indirect reference to “you” is used here instead of the official word for “you” in Bulgarian, which in this case would be “Вие”. The sentence is formal and there is no differentiation of gender in Bulgarian for the singular formal “you”. What is interesting here is that PECoRe actually takes into account the context and translates the sentence using the patterns from the context. The word in question from the context is “обмислите” which means “you have to take into consideration“ (that one word means the whole phrase in English, including “you”). The formal “you” is indirectly used and represented in the last 2 letters of the word which are underlined. This indirect “you” is picked up by PECoRe and the model translates the “you” in the output current exactly the same way, by using “-те” at the end of the words “travel” and “expect”, so the context here is taken into account and the translation matches that.

**Sentence 25**: In this example the CTI identified 3 tokens, 2 of which highlight the phenomena in this study. The source sentence here has 2 uses of the pronoun “you” and in the context is it seen once. What is interesting here is that PECoRe outputs two different types of “you” in Bulgarian. The first one is the formal singular form of “you” = “вас” which is the official word in Bulgarian and has not been used in the previous examples above. It derives it from the context, more specifically the word “сте” which is an indirect reference to “you”. So, PECoRe takes into account the indirectly inferred context and translates it into the direct word for “you”. It is not clear why it changes it, even though it is still a correct translation. The second “you”, however, is indirect, as described in the previous sentences and PECoRe relies again on the same word “сте”. So in this example, PECoRe takes into account the context and outputs two versions of “you” even though only one of them is shown in the context.

**Sentence 29**: There are two generated outputs here with the second one being related to pronouns. There are 4 uses of “you” and 1 use of “your” in the output sentence in Bulgarian. PECoRe takes into consideration the context where the “you” is indirect and translates it in the exact same way. It does the same for “your” which is represented by the word “си” in Bulgarian and is, again, indirect form rather than the official word for “your”. However, PECoRe is consistent with the translation here and relies on the seen context, using the same indirect form of “you” and “your”.

**Sentence 39**: There are 6 generated outputs here, the fourth one being the only relevant for this analysis. In the source sentence there are 3 uses of “you” and in the context there are 2 uses of “you”. Here the context in Bulgarian uses the informal singular form of “you” in an indirect way by the use of “-ш” at the end of the verb. PECoRe picks up on that and uses it in the generated translation as all the verbs also end in “-ш”. In English “you” remains the same whether it is formal or informal, however, in Bulgarian that is not the case and often the choice depends on the context. Here, the context matters because it uses the informal form, therefore the translation is consistent with that. PECoRe outputs the word “теб” as well which technically means “you” but more specifically, it means “with you” and again, the informal form is used even though that word is not present in the context. However, since the context is informal throughout, PECoRe makes a preference for the informal form in the translation too.

**Sentence 53**: There are 5 generated outputs here and neither one highlights the phenomena in question.
