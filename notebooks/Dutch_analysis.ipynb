# **Analysis of the 6 common sentences across all sentences**
**Sentence 12**: The CTI  highlights the pronoun uw we are looking for (with a high sensitivity) and is the formal translation of your. The CCI emphasises requirement and . in the input context and the . in the output context as the tokens to provide the contextual cues. However, these tokens do not express formality at all. Both contexts are written in a formal manner, intuitively triggering the formal translation. PECoRe did not pick up on these tokens that have formality attached to them, such as: vereist, legaliseren and consulaat.

**Sentence 18**: The CTI highlights the desired token je with a high sensitivity. The CCI displays that the token je in the output context provided the contextual cue, which is correct. Notably, the second instance of je was not marked by the CTI as context-sensitive. Given the reliance of the first instance, we presume the second instance of je in the output relies on the second instance of je in the output context.

**Sentence 25**: Similar to sentence 18, the first instance of je is tagged by the CTI, although with a lower sensitivity, and the second instance is not marked as context-sensitive. Again, je relies on je in the output context. In addition, the CTI did mark heb as sensitive to the output context  je, which is correct from a grammatical perspective.

**Sentence 29**: Surprisingly, all four instances of je are not highlighted as context-sensitive. Additionally, the last instance of je maps to your, for which jouw is more expected. Note that jouw is formal and high-level informal, the use of je for your is very informal. The translated context consistently used je for both you and your, which intuitively led to the consistent use of je for both pronouns.

**Sentence 39**: In agreement with the findings of sentences 18 and 25, the CTI only highlighted the first instance of je (low sensitivity) as context-sensitive. The other two instances of je are not marked by the CTI as context-sensitive. The CCI discloses the tokens Wanneer and je in the output context provided the contextual cues, of which only je is correct. The token wanneer does not express formality and is therefore unreasonable to use. Moreover but unrelated to formality, the CTI and CCI beautifully mapped mug to mug in the output context.

**Sentence 53**: There is no form of you, your or yours present in the output current nor the output context. Nonetheless, three outputs are generated. The translation (output current) is correct, but does not involve the phenomenon of interest.



# **Analysis of the remaining 5 sentences**

**Sentence 3**: One generated output (see above). Surprisingly, the pronoun u is not marked as context-sensitive by the CTI. Although, the full stop (.) is marked as highly context-sensitive.
